# Sampling error in inferential statistics

Since the size of a sample is always smaller than the size of the population, some of the population isnâ€™t captured by sample data. This creates sampling error, which is the difference between the true population values (called parameters) and the measured sample values (called statistics).

Sampling error arises any time you use a sample, even if your sample is random and unbiased. For this reason, there is always some uncertainty in inferential statistics. However, using probability sampling methods reduces this uncertainty.




# Estimating population parameters from sample statistics

The characteristics of samples and populations are described by numbers called statistics and parameters:

A statistic is a measure that describes the sample (e.g., sample mean).
A parameter is a measure that describes the whole population (e.g., population mean).
Sampling error is the difference between a parameter and a corresponding statistic. Since in most cases you donâ€™t know the real population parameter, you can use inferential statistics to estimate these parameters in a way that takes sampling error into account.




There are two important types of estimates you can make about the population: point estimates and interval estimates.

- A point estimate is a single value estimate of a parameter. For instance, a sample mean is a point estimate of a population mean.
- An interval estimate gives you a range of values where the parameter is expected to lie. A confidence interval is the most common type of interval estimate.

Both types of estimates are important for gathering a clear idea of where a parameter is likely to lie.




# Confidence intervals

A confidence interval uses the variability around a statistic to come up with an interval estimate for a parameter. Confidence intervals are useful for estimating parameters because they take sampling error into account.

While a point estimate gives you a precise value for the parameter you are interested in, a confidence interval tells you the uncertainty of the point estimate. They are best used in combination with each other.

Each confidence interval is associated with a confidence level. A confidence level tells you the probability (in percentage) of the interval containing the parameter estimate if you repeat the study again.

A 95% confidence interval means that if you repeat your study with a new sample in exactly the same way 100 times, you can expect your estimate to lie within the specified range of values 95 times.

Although you can say that your estimate will lie within the interval a certain percentage of the time, you cannot say for sure that the actual population parameter will. Thatâ€™s because you canâ€™t know the true value of the population parameter without collecting data from the full population.

However, with random sampling and a suitable sample size, you can reasonably expect your confidence interval to contain the parameter a certain percentage of the time.

# Calculating a Confidence Interval

You survey 100 Brits and 100 Americans about their television-watching habits, and find that both groups watch an average of 35 hours of television per week.
However, the British people surveyed had a wide variation in the number of hours watched, while the Americans all watched similar amounts.

Even though both groups have the same point estimate (average number of hours watched), the British estimate will have a wider confidence interval than the American estimate because there is more variation in the data.

## Components Required

1. **Point Estimate**: This is the statistical estimate you're making (e.g., population mean).
2. **Critical Values**: These values depend on the desired confidence level for your CI.
3. **Standard Deviation (s)**: The spread of your sample data.
4. **Sample Size (n)**: The number of observations in your sample.

## Steps to Calculate Confidence Interval

### 1. Point Estimate Example

For a study on TV-watching habits, the point estimate might be the mean number of hours watched, say 35 hours.

### 2. Finding the Critical Value

- **Alpha (Î±) Value**: Commonly set at p = 0.05 for a 95% CI.
- **One-Tailed vs. Two-Tailed**: Typically, you'll use a two-tailed interval.
- **Critical Value for Z Distribution**: For a 95% CI, the z statistic is 1.96.

If your data follows a normal distribution, or if you have a large sample size (n > 30) that is approximately normally distributed, you can use the z distribution to find your critical values.

For a z statistic, some of the most common values are shown in this table:



If you are using a small dataset (n â‰¤ 30) that is approximately normally distributed, use the t distribution instead.

Check out this set of [t tables](https://www.scribbr.com/statistics/students-t-table/) to find your t statistic. We have included the confidence level and p values for both one-tailed and two-tailed tests to help you find the t value you need.

### 3. Standard Deviation and Sample Size

- **To Find Standard Deviation**: Calculate the sample variance and take its square root.
- **Sample Size**: The total number of observations in your study.

## Confidence Interval Formula

For normally-distributed data, the confidence interval is calculated as:

$$CI = \hat{x} \pm Z^* \frac{s}{\sqrt{n}}$$

- $\hat{x}$ = Sample mean
- $Z^*$ = Critical value from the z distribution
- $s$ = Sample standard deviation
- $\sqrt{n}$ = Square root of the sample size

### Example: Calculating the Confidence Interval

- **For the USA**: With a sample mean of 35, a standard deviation of 5, and a sample size of 100:

  $$CI = 35 \pm 1.96 \frac{5}{\sqrt{100}} = 35 \pm 0.98$$

  The 95% confidence interval is 34.02 to 35.98.

- **For GB**: With the same mean but a standard deviation of 10:

  $$CI = 35 \pm 1.96 \frac{10}{\sqrt{100}} = 35 \pm 1.96$$

  The 95% confidence interval is 33.04 to 36.96.


# Hypothesis testing

Hypothesis testing is a process to assess relationships or compare populations using samples. It involves statistical tests, which can be parametric (assuming a normal distribution and other conditions) or non-parametric (not assuming a specific distribution).

Statistical tests come in three forms: tests of comparison, correlation or regression.

# Statistical tests

there is a various statistical tests for comparison (e.g., t-test, ANOVA), correlation (e.g., Pearsonâ€™s r, Spearmanâ€™s r), and regression (e.g., linear regression, logistic regression), each suited for different types of data and research questions.

## Comparison tests

Comparison tests assess whether there are differences in means, medians or rankings of scores of two or more groups.



## Correlation tests

Correlation tests determine the extent to which two variables are associated.



## Regression tests

Regression tests demonstrate whether changes in predictor variables cause changes in an outcome variable. You can decide which regression test to use based on the number and types of variables you have as predictors and outcomes.





# Example

### Scenario

A school has introduced a new teaching method aimed at improving student performance in mathematics. To evaluate the effectiveness of this method, we'll compare the test scores of students taught using the new method to those taught using the traditional method.

First, we collect data from a sample of students. For simplicity, let's assume we've gathered test scores from 30 students taught by the new method and 30 students taught by the traditional method.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Generating synthetic data
np.random.seed(0)
new_method_scores = np.random.normal(loc=75, scale=10, size=30)  # New teaching method
traditional_method_scores = np.random.normal(loc=70, scale=15, size=30)  # Traditional method

# Descriptive statistics
print(f"New Method: Mean = {new_method_scores.mean():.2f}, SD = {new_method_scores.std():.2f}")
print(f"Traditional Method: Mean = {traditional_method_scores.mean():.2f}, SD = {traditional_method_scores.std():.2f}")
```

```python
# Hypothesis Testing

#To determine if the new teaching method significantly impacts student performance, 
#we'll conduct a hypothesis test (t-test) since we're comparing the means of two independent samples.

# Conducting a t-test
t_stat, p_value = stats.ttest_ind(new_method_scores, traditional_method_scores)

print(f"T-Statistic: {t_stat:.2f}, P-value: {p_value:.3f}")
```

```python

mean_new = np.mean(new_method_scores)
mean_trad = np.mean(traditional_method_scores)
std_new = np.std(new_method_scores, ddof=1)
std_trad = np.std(traditional_method_scores, ddof=1)


n_new = len(new_method_scores)
n_trad = len(traditional_method_scores)

# Calculate the standard error of the difference between the means
sed = np.sqrt((std_new**2 / n_new) + (std_trad**2 / n_trad))

# Calculate the t statistic for a 95% confidence interval
# This uses the degrees of freedom for two independent samples
df = min(n_new - 1, n_trad - 1)
t_critical = stats.t.ppf(0.975, df)  # 0.975 because 2-tailed, and we're interested in the 95% confidence interval

# Calculate the margin of error
margin_of_error = t_critical * sed

# Calculate the confidence interval
ci_lower = (mean_new - mean_trad) - margin_of_error
ci_upper = (mean_new - mean_trad) + margin_of_error

print(f"95% Confidence Interval for the mean difference: ({ci_lower:.3f}, {ci_upper:.3f})")
```

In our scenario, we used a t-test for comparing means between two independent groups, which is suitable for our normally distributed interval data

```python
plt.hist(new_method_scores, alpha=0.5, label='New Method')
plt.hist(traditional_method_scores, alpha=0.5, label='Traditional Method')
plt.legend(loc='upper right')
plt.xlabel('Scores')
plt.ylabel('Frequency')
plt.title('Distribution of Test Scores')
plt.show()
```
