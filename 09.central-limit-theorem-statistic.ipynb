The central limit theorem states that if you take sufficiently large samples from a population, the samples’ means will be normally distributed, even if the population isn’t normally distributed.

## Example

A population follows a Poisson distribution (left image). If we take 10,000 samples from the population, each with a sample size of 50, the sample means follow a normal distribution, as predicted by the central limit theorem (right image).




# ***What is the central limit theorem?***

The central limit theorem relies on the concept of a sampling distribution, which is the probability distribution of a statistic for a large number of samples taken from a population.

Imagining an experiment may help you to understand sampling distributions:

- Suppose that you draw a random sample from a population and calculate a statistic for the sample, such as the mean.
- Now you draw another random sample of the same size, and again calculate the mean.
- You repeat this process many times, and end up with a large number of means, one for each sample.

The distribution of the sample means is an example of a sampling distribution.

The central limit theorem says that the sampling distribution of the mean will always be normally distributed, as long as the sample size is large enough. Regardless of whether the population has a normal, Poisson, binomial, or any other distribution, the sampling distribution of the mean will be normal.

A normal distribution is a symmetrical, bell-shaped distribution, with increasingly fewer observations the further from the center of the distribution.

# Sample size and the central limit theorem


The impact of sample size on the sampling distribution of the mean can be summarized through two main points:

## 1.Sample Size and Normality:

- A larger sample size makes the sampling distribution more closely follow a normal distribution, aligning with the central limit theorem. This theorem holds when the sample size is considered "sufficiently large," usually 30 or more observations.

- If the sample size is less than 30, the central limit theorem may not apply, and the sampling distribution's shape will mirror that of the population. This means it will only be normal if the population distribution is normal.

- For sample sizes of 30 or more, the sampling distribution is approximately normal regardless of the population's distribution.

## 2.Sample Size and Standard Deviations:

The standard deviation of the sampling distribution, which measures the variability or spread of the sample means, is inversely related to the sample size.

- A smaller sample size results in a higher standard deviation, indicating a wider spread in the sampling distribution due to less precise estimates of the population mean.
- Conversely, a larger sample size leads to a lower standard deviation, reflecting a narrower spread and more precise estimates of the population mean.

In summary, increasing the sample size leads to a more normally distributed sampling distribution and a reduction in the standard deviation of the sampling distribution, thereby providing more precise estimates of the population mean.

# Example:

```python
import numpy as np
import matplotlib.pyplot as plt

#  Poisson distribution
lambda_ = 3 
population_size = 10000

#
population = np.random.poisson(lambda_, population_size)


sample_size = 50
number_of_samples = 10000


sample_means = [np.mean(np.random.choice(population, sample_size, replace=False)) for _ in range(number_of_samples)]


fig, axs = plt.subplots(1, 2, figsize=(14, 7))

#  population distribution
axs[0].hist(population, bins=30, color='blue', alpha=0.7)
axs[0].set_title('Population Distribution\n(Poisson Distribution)')
axs[0].set_xlabel('Value')
axs[0].set_ylabel('Frequency')

#  distribution of sample means
axs[1].hist(sample_means, bins=30, color='orange', alpha=0.7)
axs[1].set_title('Distribution of Sample Means\n(Normal Distribution)')
axs[1].set_xlabel('Sample Mean')
axs[1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```
