| Distribution Type | Distribution        | Definition                                                                                                                                              | Use-Cases                                                                                                                                                    | Notation                        | Mean                                 | Variance                              |
|-------------------|---------------------|---------------------------------------------------------------------------------------------------------------------------------------------------------|-------------------------------------------------------------------------------------------------------------------------------------------------------------|---------------------------------|--------------------------------------|---------------------------------------|
| Discrete          | Binomial            | Number of successes in a fixed number of independent trials with a constant probability of success.                                                     | Counting successes in independent and identically distributed Bernoulli trials.                                                                              | $$ B(n, p) $$                   | $$ np $$                             | $$ np(1-p) $$                         |
| Discrete          | Poisson             | Probability of a given number of events occurring in a fixed interval of time or space, with a known constant mean rate and independent of the last event. | Modeling the number of events in a time or space interval.                                                                                                   | $$ \text{Pois}(\lambda) $$      | $$ \lambda $$                        | $$ \lambda $$                         |
| Discrete          | Discrete Uniform    | Each of the n outcomes has an equal probability.                                                                                                        | Scenarios where each outcome in a finite set is equally likely.                                                                                              | $$ U(a, b) $$                   | $$ \frac{a+b}{2} $$                  | $$ \frac{(b-a+1)^2 - 1}{12} $$        |
| Discrete          | Bernoulli           | Only two possible outcomes, success (1) and failure (0), in a single trial.                                                                             | Events with two possible outcomes (e.g., success/failure).                                                                                                   | $$ B(p) $$                      | $$ p $$                              | $$ p(1-p) $$                          |
| Continuous        | Normal              | Symmetric, bell-shaped distribution characterized by its mean and standard deviation.                                                                    | Measurements clustering around a single mean value.                                                                                                           | $$ N(\mu, \sigma^2) $$          | $$ \mu $$                            | $$ \sigma^2 $$                        |
| Continuous        | Uniform             | All outcomes are equally likely across a range.                                                                                                         | Events with equal chances of occurring across a continuous range.                                                                                            | $$ U(a, b) $$                   | $$ \frac{a+b}{2} $$                  | $$ \frac{(b-a)^2}{12} $$              |
| Continuous        | Exponential         | Time between events in a Poisson process, where events occur continuously and independently at a constant average rate.                                  | Time until a continuous process occurs (e.g., time between failures).                                                                                        | $$ \text{Exp}(\lambda) $$       | $$ \frac{1}{\lambda} $$             | $$ \frac{1}{\lambda^2} $$             |
| Continuous        | Student's t         | Like the normal distribution but with heavier tails, indicating a higher likelihood of extreme values.                                                   | Small sample size data analysis when the population variance is unknown.                                                                                     | $$ t(df) $$                     | $$ 0 \text{ if } df > 1 $$           | $$ \frac{df}{df-2} \text{ if } df > 2 $$ |
| Continuous        | Chi-Square          | Sum of the squares of k independent standard normal random variables.                                                                                    | Hypothesis testing and confidence intervals for variance and standard deviations when the distribution is normal.                                            | $$ \chi^2(k) $$                 | $$ k $$                              | $$ 2k $$                              |
