The standard error of the mean, or simply standard error, indicates how different the population mean is likely to be from a sample mean. It tells you how much the sample mean would vary if you were to repeat a study using new samples from within a single population.

The standard error of the mean (SE or SEM) is the most commonly reported type of standard error. But you can also find the standard error for other statistics, like medians or proportions. The standard error is a common measure of sampling errorâ€”the difference between a population parameter and a sample statistic.


Standard error measures how well a sample represents a population, indicating the spread of sample means around the population mean. A high standard error suggests the sample may not closely represent the population, while a low standard error indicates a more representative sample. Sampling error is inevitable, but can be minimized by increasing the sample size. Using large, random samples is crucial for reducing sampling bias and improving the accuracy of conclusions about the population

# Standard error vs standard deviation

Standard error and standard deviation are both measures of variability:

- The standard deviation describes variability within a single sample.
- The standard error estimates the variability across multiple samples of a population.



The standard deviation is a descriptive statistic that can be calculated from sample data. In contrast, the standard error is an inferential statistic that can only be estimated (unless the real population parameter is known).

# How should you report the standard error?

You can report the standard error alongside the mean or in a confidence interval to communicate the uncertainty around the mean.

The best way to report the standard error is in a confidence interval because readers wonâ€™t have to do any additional math to come up with a meaningful interval.

A confidence interval is a range of values where an unknown population parameter is expected to lie most of the time, if you were to repeat your study with new random samples.

With a 95% confidence level, 95% of all sample means will be expected to lie within a confidence interval of Â± 1.96 standard errors of the sample mean.

Based on random sampling, the true population parameter is also estimated to lie within this range with 95% confidence.
